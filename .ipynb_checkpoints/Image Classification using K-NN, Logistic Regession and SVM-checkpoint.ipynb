## Part 1. k-NN Classifier
For this section we will implement a simple kNN Classifier on the MNIST dataset. To do this we will need to perform the following steps:
- Create a k-NN class. Our class must contain a method that returns predictions for our test set.
- Provide a suitable distance metric that we will use to calculate the nearest neightbours. We may choose the distance metric we believe is most suitable.
- Calculate the k nearest neighbours and make predictions.
- Choose the a value for _k_ that results in the highest accuracy on our test set.

When our classifier is working:
- Use sklearn to calculate accuracy and plot a confusion matrix using our predictions.
- A brief discussion of our results

### Note:

When we compute the distance metric we can acheive better performance if we vectorize the computation instead of using for loops to iterate through the values.




````python
from math import sqrt
from tqdm import tqdm
from collections import Counter

class KNeighborsClassifier:
    def __init__(self, n_neighbors=5):
        """Classifier implementing the k-nearest neighbors vote.

        Parameters
        ----------
        n_neighbors : int, default=5
            Number of neighbors to use by default for :meth:`kneighbors` queries.
        """
        self.n_neighbors = n_neighbors

    def fit(self, X, y):
        """Set the values for training data and training labels.

        Parameters
        ----------
        X : {array-like, sparse matrix} of shape (n_samples, n_features) or \
                (n_samples, n_samples)
            Training data.
        y : {array-like, sparse matrix} of shape (n_samples,) or \
                (n_samples, n_outputs)
            Training labels.
        """
        self.X = X
        self.y = y

        return self

    def euclidean_distance(self, x1, x2):
        """Calculate Euclidean distance between two points.

        Parameters
        ----------
        x1 : numpy array
            1st point.
        x2 : numpy array
            2nd point.
        """
        return np.sqrt(np.sum(np.square(x1 - x2)))

    def get_neighbors(self, test_sample):
        """Return the K-NN.

        Parameters
        ----------
        test_sample : {array-like, sparse matrix} of shape (n_samples,) or \
                (n_samples, n_outputs)
            Test sample.
        """
        distances = list()
        index = 0
        for train_sample in self.X:
            distance = self.euclidean_distance(train_sample, test_sample)
            distances.append((index, distance))
            index += 1

        # Sort distances by distance
        distances.sort(key=lambda record: record[1])
        neighbors = list()
        for i in range(self.n_neighbors):
            neighbors.append(distances[i][0])

        return neighbors

    def predict_sample(self, test_sample):
        """Return the prediction of a given sample.

        Parameters
        ----------
        test_sample : {array-like, sparse matrix} of shape (n_samples,) or \
                (n_samples, n_outputs)
            Test sample.
        """
        neighbors = self.get_neighbors(test_sample)
        labels = [self.y[i] for i in neighbors]
        prediction = max(set(labels), key=labels.count)

        return prediction

#     def predict(self, X_test):
#         """Return the prediction of a given test dataset.

#         Parameters
#         ----------
#         X_test : {array-like, sparse matrix} of shape (n_samples,) or \
#                 (n_samples, n_outputs)
#             Test dataset.
#         """
#         predictions = list()
#         for i in tqdm(range(X_test.shape[0])):
#             prediction = self.predict_sample(X_test[i])
#             predictions.append(prediction)
        
#         return predictions

    def get_distances(self, X_test):
        """Calculate Euclidean distances between two matrices.

        Parameters
        ----------
        X_test : {array-like, sparse matrix} of shape (n_samples, n_features) or \
                (n_samples, n_samples)
            Testing data.
        """
        distances = np.zeros((X_test.shape[0], self.X.shape[0]))

        X2 = self.X * self.X
        X2 = np.sum(X2, axis = 1)
        X2_repeat = np.array([X2] * X_test.shape[0])

        X_test2 = X_test * X_test
        X_test2 = np.sum(X_test2, axis = 1)
        X_test2_repeat = np.array([X_test2] * self.X.shape[0]).transpose()

        X_test_dot_X = X_test.dot(self.X.T)

        distances = X2_repeat + X_test2_repeat - 2 * X_test_dot_X
        distances = np.sqrt(distances)

        return distances

    def predict_labels(self, distances):
        """Return the prediction labels.

        Parameters
        ----------
        distances : {array-like, sparse matrix} of shape (n_samples,) or \
                (n_samples, n_outputs)
            Distances matrix.
        """
        y_pred = np.zeros(distances.shape[0]).astype("int")
        for index in range(distances.shape[0]):
            distances_i = distances[index]
            knn = self.y[distances_i.argsort()[:self.n_neighbors]]
            y_pred[index] = Counter(knn).most_common(1)[0][0]

        return y_pred

    def predict(self, X_test):
        """Return the prediction of a given test dataset.

        Parameters
        ----------
        X_test : {array-like, sparse matrix} of shape (n_samples,) or \
                (n_samples, n_outputs)
            Test dataset.
        """
        distances = self.get_distances(X_test)
        return self.predict_labels(distances)
````

````python
model = KNeighborsClassifier(n_neighbors=3)
predictions = model.fit(train_images, train_labels).predict(test_images)
````

````python
accuracy = float(np.sum(predictions == test_labels)) / test_images.shape[0] * 100
print(f"Accuracy: {accuracy}%")
````

## Part 2. Logistic Regression

For this section we will perform binary classification using logistic regression. Just as in Part 1. we will use the MNIST dataset, however to obtain a result for each class using logistic regression we will need to use a One-vs-Rest (OvR) approach to acheive multi-class classification. 

We will need to follow these basic steps:
- Train a binary classifier for each class, where the target class is a "positive" results and the combination of the remaining classes are "negative". For MNIST we will need 10 models.
- For each test sample compute the probabilities for each model
- Select the argmax of the probabilities to obtain the predicted class

Collect your predictions from the test set and compute the accuracy score and plot a confusion matrix.

````python
def get_estimators(estimator):
    estimators = list()
    
    for c in tqdm(range(10)):
        y_train = train_labels.copy()
        y_test = test_labels.copy()

        y_train[y_train != c] = -1
        y_train[y_train == c] = 1
        y_train[y_train == 255] = 0
        y_test[y_test != c] = -1
        y_test[y_test == c] = 1
        y_test[y_test == 255] = 0

        estimator = clone(estimator)
        estimator = estimator.fit(train_images, y_train)
        estimators.append(estimator)

    return estimators
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.base import clone

lr_estimators = get_estimators(LogisticRegression(max_iter=1000))
````

````python
def prediction(x, estimators):
    probabilities = list()
    for index, estimator in enumerate(estimators):
        prediction = estimator.predict_proba(x)
        probabilities.append(prediction[0][1])

    return np.argmax(probabilities)
````

````python
# OvR-LogisticRegression

y_pred_ovr_lr = np.zeros((test_images.shape[0], 1))
for i in range(test_images.shape[0]):
    y_pred_ovr_lr[i] = prediction(test_images[i,:].reshape(test_images.shape[1], 1).T, lr_estimators)
````

````python
target_names = [str(c) for c in set(test_labels)]
print(classification_report(test_labels, y_pred_ovr_lr, target_names=target_names, digits=10))
accuracy_score(test_labels, y_pred_ovr_lr)
````

````python
cm = confusion_matrix(test_labels, y_pred_ovr_lr)
print(cm)
````

````python
from sklearn.svm import LinearSVC

# Create the SVM
svm = LinearSVC(random_state=42, multi_class='ovr')

# Fit the data to the OvR classifier
svm = svm.fit(train_images, train_labels)

# Predictions
y_pred_ovr_svm = svm.predict(test_images)

# Evaluate by means of a confusion matrix
target_names = [str(c) for c in set(test_labels)]
print(classification_report(test_labels, y_pred_ovr_svm, target_names=target_names, digits=10))
accuracy_score(test_labels, y_pred_ovr_svm)
````

````python
cm = confusion_matrix(test_labels, y_pred_ovr_svm)
print(cm)
````

````python
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]

accuracies = list()
for i in tqdm(range(len(C))):
    svm = LinearSVC(random_state=42, multi_class='ovr', C=C[i])
    svm = svm.fit(train_images, train_labels)
    y_pred_ovr_svm = svm.predict(test_images)
    accuracies.append(accuracy_score(test_labels, y_pred_ovr_svm))
````

````python
plt.figure(figsize=(10, 6))
plt.plot(C, accuracies, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('C vs. Accuracy')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.xscale('log')
for i in range(len(C)):
    plt.annotate(C[i], (C[i], accuracies[i] - 0.003))
plt.show()
````

````python
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]

accuracies = list()
for i in tqdm(range(len(C))):
    svm = LinearSVC(random_state=42, multi_class='ovr', C=C[i], penalty='l1', dual=False)
    svm = svm.fit(train_images, train_labels)
    y_pred_ovr_svm = svm.predict(test_images)
    accuracies.append(accuracy_score(test_labels, y_pred_ovr_svm))
````

````python
plt.figure(figsize=(10, 6))
plt.plot(C, accuracies, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('C vs. Accuracy')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.xscale('log')
for i in range(len(C)):
    plt.annotate(C[i], (C[i], accuracies[i] - 0.003))
plt.show()
````

````python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

accuracies = list()
for cost in [0.01, 0.1, 1, 10, 100]:
    for gamma in [0.01, 0.1, 1, 10, 100]:
        svm = SVC(random_state=42, kernel='rbf', C=cost, gamma=gamma)
        svm = OneVsRestClassifier(svm)
        svm = svm.fit(train_images[:10000], train_labels[:10000])
        y_pred_ovr_svm = svm.predict(test_images[:2000])
        accuracies.append((cost, gamma, accuracy_score(test_labels[:2000], y_pred_ovr_svm[:2000])))
        print((cost, gamma, accuracy_score(test_labels[:2000], y_pred_ovr_svm[:2000])))
````

````python
plt.figure(figsize=(10, 6))
plt.plot(tuple(zip(*accuracies))[0], tuple(zip(*accuracies))[1], color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('C vs. gamma')
plt.xlabel('C')
plt.ylabel('gamma')
plt.xscale('log')
plt.yscale('log')
for i in range(len(accuracies)):
    plt.annotate(round(tuple(zip(*accuracies))[2][i]*100, 2), (tuple(zip(*accuracies))[0][i] + 0.003, tuple(zip(*accuracies))[1][i] + 0.004))
plt.show()
````

````python
# As we can see the highest accuracy obtained at {C=10, gamma=.01} and {C=100, gamma=0.01}
# Train classifiers for those 2 cases on all data
accuracies = list()
for cost in [10, 100]:
    for gamma in [0.01]:
        svm = SVC(random_state=42, kernel='rbf', C=cost, gamma=gamma)
        svm = OneVsRestClassifier(svm)
        svm = svm.fit(train_images, train_labels)
        y_pred_ovr_svm = svm.predict(test_images)
        accuracies.append((cost, gamma, accuracy_score(test_labels, y_pred_ovr_svm)))
        print((cost, gamma, accuracy_score(test_labels, y_pred_ovr_svm)))
````

````python
# The model with the highest accuracy

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# As we can see the highest accuracy obtained at {C=100, gamma=0.01}
svm = SVC(random_state=42, kernel='rbf', C=100, gamma=0.01)
svm = OneVsRestClassifier(svm)
svm = svm.fit(train_images, train_labels)
y_pred_ovr_svm = svm.predict(test_images)

# Evaluate by means of a confusion matrix
target_names = [str(c) for c in set(test_labels)]
print(classification_report(test_labels, y_pred_ovr_svm, target_names=target_names, digits=10))
accuracy_score(test_labels, y_pred_ovr_svm)
````

### Discussion

- As we can see from the results obtained above, the baseline accuracy is 91.84%, after trying multilpe cost and gamma values, we obtained much higher accuracy; 98.52% which is a major enhancement in the model accuracy, we obtained this results using cost=100 and gamma=0.01

- According to theory, prediction consistency is not possible because of the bias of the l1. However, model consistency can be achieved by scaling C1. As we can see increasing C while we have l1 makes our predictions consistent

- C behaves as a regularization parameter in the SVM, as we can see when increasing the parameter C, a smaller margin will be accepted if the decision function is better at classifying all training points correctly. Higher C values lead to less support vectors and therefore less prediction time. Low values of gamma means far influence of training examples; we used gamma=0.01 in our code.

- As we can see from the obtained accuracies of the model on training dataset and testing dataset, no overfitting is observed since our model doesn't fail to generalize well, in fact the model has very high accuracy on both training and testing datasets.

- As we can see from the above confusion matrix, only few misclassified examples exist in different classes, mostly because the data is not separable and there exist a little overlap.
